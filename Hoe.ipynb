```python
import matplotlib.pyplot as plt
```

#### Constants

```python
hand_time = 0.75
wood_time = 0.4
stone_time = 0.2
iron_time = 0.15
gold_time = 0.1
craft_hoe_time = 7
hand_durability = 9999
wood_durability = 59
stone_durability = 131
iron_durability = 250
gold_durability = 32
```

```python
def hay_broken_over_time(total_time, time_per_hay, durability, craft_time):
    t = 0
    progress = 0
    progresses = []
    uses = 0
    while t < craft_time:
        t += 1
        progresses.append(0)
    while t < total_time and uses < durability:
        progresses.append(progress)
        progress += min( (1 / time_per_hay, durability - uses) )
        uses += 1/ time_per_hay
        t += 1
    while t < total_time:
        progresses.append(progress)
        progress += 1 / hand_time
        t += 1
    return progresses
```

```python
t = 18
hand_values = hay_broken_over_time(t, hand_time, hand_durability, 0)
wood_values = hay_broken_over_time(t, wood_time, wood_durability, craft_hoe_time)
stone_values = hay_broken_over_time(t, stone_time, stone_durability, craft_hoe_time)
iron_values = hay_broken_over_time(t, iron_time, iron_durability, craft_hoe_time)
gold_values = hay_broken_over_time(t, gold_time, gold_durability, craft_hoe_time)
x = range(t)

# Plotting all graphs
fig = plt.figure(figsize=(16, 8))
plt.plot(x, hand_values, label='Hand', lw=6, color=(0.1, 0.7, 0.85))
plt.plot(x, wood_values, label='Wood', lw=6, color=(0.7, 0.5, 0.15))
plt.plot(x, stone_values, label='Stone', lw=6, color=(0.7, 0.7, 0.7))
plt.plot(x, iron_values, label='Iron', lw=6, color=(0.9, 0.92, 0.85))
plt.plot(x, gold_values, label='Gold', lw=6, color=(0.99, 0.9, 0.35))
plt.legend(fontsize=20)
plt.xlabel("Time (seconds)", fontsize=25, color=(1,1,1))
plt.ylabel("Number of Hay Bales Broken", fontsize=22, color=(1,1,1))
plt.xlim(0, t-1)

ax = plt.gca()
ax.set_facecolor((0.80, 0.89, 0.89))
plt.grid(color=(1, 1, 1), lw=2)
plt.xticks(fontsize=18, color=(1,1,1))
plt.yticks(fontsize=18, color=(1,1,1))
fig.patch.set_facecolor('black')

plt.show()
```
