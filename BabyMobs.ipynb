```python
FULL_GROWING_TIME = 20 * 60 # 20 minutes
PERCENT_PER_FOOD = 10       # Feeding a baby mob reduces the time by 10%
FOOD_PER_SECOND = 5         # Holding right-click feeds the mob 5x per second
```

```python
timeRemaining = FULL_GROWING_TIME
timeElapsed = 0
timeBetweenFood = 1. / FOOD_PER_SECOND
print(timeBetweenFood)
foodUsed = 0
times = []
timesRemaining = []

while timeRemaining > 0:
    # Add values to lists for making a plot afterwards
    times.append(timeElapsed)
    timesRemaining.append(timeRemaining / 60.)
    
    # 0.2 seconds pass
    timeRemaining -= timeBetweenFood
    # The mob is fed, reducing by 10%
    timeRemaining -= (timeRemaining * (PERCENT_PER_FOOD / 100.))

    # Update the counting variables
    timeElapsed += timeBetweenFood
    foodUsed += 1

print("It took %f seconds to make the baby mob grow up." % timeElapsed)
print("This required feeding %d items." % foodUsed)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(16, 8))
plt.plot(times, timesRemaining, lw=5)

ax = plt.gca()
ax.set_facecolor((0.80, 0.89, 0.89))
plt.grid(color=(1, 1, 1), lw=2)
plt.xticks(fontsize=18, color=(1,1,1))
plt.yticks(fontsize=18, color=(1,1,1))
fig.patch.set_facecolor('black')
plt.xlim(0, 13)
plt.ylabel("Time Remaing for the Mob\n to Grow Up (Minutes)", fontsize=24, color=(1,1,1))
plt.xlabel("Time That Has Passed (Seconds)", fontsize=24, color=(1,1,1))

plt.show()
```
