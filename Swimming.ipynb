```python
import matplotlib.pyplot as plt
import numpy as np
```

### Data

```python
swim_times = {}
swim_times[0] = 0
swim_times[25] = 6.45
swim_times[50] = 12.64
swim_times[75] = 18.98
swim_times[100] = 25.43
swim_times[125] = 31.71
swim_times[150] = 38.13
swim_times[175] = 44.41
swim_times[200] = 50.78
```

```python
walk_times = {}
walk_times[0] = 0
walk_times[25] = 5.97
walk_times[50] = 11.76
walk_times[75] = 17.65
walk_times[100] = 23.46
walk_times[125] = 29.30
walk_times[150] = 35.07
walk_times[175] = 40.82
walk_times[200] = 46.60
```

```python
run_times = {}
run_times[0] = 0
run_times[25] = 4.67
run_times[50] = 9.17
run_times[75] = 13.61
run_times[100] = 18.06
run_times[125] = 22.56
run_times[150] = 26.93
run_times[175] = 31.49
run_times[200] = 35.88
```

## Graphs

```python
fig = plt.figure(figsize=(16, 8))
plt.xlabel("Time (seconds)", fontsize=25, color=(1,1,1))
plt.ylabel("Distance Traveled (meters)", fontsize=25, color=(1,1,1))
plt.plot(swim_times.values(), swim_times.keys(), lw=5, label="Swimming", color=(0.2, 0.8, 1))
plt.plot(walk_times.values(), walk_times.keys(), lw=5, label="Walking", color=(1, 1, 0.1))
plt.plot(run_times.values(), run_times.keys(), lw=5, label="Sprinting", color=(1, 0.4, 0.85))
plt.legend(fontsize=24)
plt.xlim(0, 51)
plt.ylim(0, 200)

ax = plt.gca()
ax.set_facecolor((0.85, 0.85, 0.95))
plt.grid(color=(1, 1, 1), lw=2)
plt.xticks(fontsize=18, color=(1,1,1))
plt.yticks(fontsize=18, color=(1,1,1))
fig.patch.set_facecolor('black')

plt.show()
```

### Lines of Best Fit

```python
x = np.array(list(swim_times.values()))
y = np.array(list(swim_times.keys()))
m, b = np.polyfit(x, y, 1)
print("Swim speed is calculated to be {:.2f} m/s".format(m))
```

```python
x = np.array(list(walk_times.values()))
y = np.array(list(walk_times.keys()))
m, b = np.polyfit(x, y, 1)
print("Walk speed is calculated to be {:.2f} m/s".format(m))
```

```python
x = np.array(list(run_times.values()))
y = np.array(list(run_times.keys()))
m, b = np.polyfit(x, y, 1)
print("Sprint speed is calculated to be {:.2f} m/s".format(m))
```
