```python
import matplotlib.pyplot as plt
import numpy as np
```

### Constants

```python
swim_speed = 5.612
boat_speed = 8
craft_time = 12
```

```python
def swim_distance(total_time):
    distance = 0
    distances = []
    for t in range(total_time):
        distances.append(distance)
        distance += swim_speed
    return distances
```

```python
def boat_distance(total_time):
    t = 0
    distance = 0
    distances = []
    while t < craft_time:
        t += 1
        distances.append(distance)
    while t < total_time:
        distances.append(distance)
        distance += boat_speed
        t += 1
    return distances
```

```python
t = 61
swim_values = swim_distance(t)
boat_values = boat_distance(t)
x = range(t)

# Plotting all graphs
fig = plt.figure(figsize=(16, 8))
plt.plot(x, swim_values, label='Swim', lw=5, color=(0.1, 0.9, 1))
plt.plot(x, boat_values, label='Boat', lw=5, color=(1, 0.4, 0.85))
plt.legend(fontsize=20)
plt.xlabel("Time (seconds)", fontsize=25, color=(1,1,1))
plt.ylabel("Distance Traveled", fontsize=22, color=(1,1,1))
plt.xlim(0, t-1)

ax = plt.gca()
ax.set_facecolor((0.85, 0.85, 0.95))
plt.grid(color=(1, 1, 1), lw=2)
plt.xticks(fontsize=18, color=(1,1,1))
plt.yticks(fontsize=18, color=(1,1,1))
fig.patch.set_facecolor('black')

plt.show()
```
